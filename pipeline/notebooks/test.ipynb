```python
# Hacemos un pca
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Excluimos la columna de 'cluster'
X = data
# Convertimos a numpy array
X = X.to_numpy()
# Vemos el tamaÃ±o de los datos
print(f"Shape of data: {X.shape}")
# Estandarizamos los datos
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
```

```python
# Aplicamos el PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)
# Vemos el tamaÃ±o de los datos transformados
print(f"Shape of PCA data: {X_pca.shape}")
# Vemos la varianza explicada por cada componente
print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
# Vemos la varianza total explicada acumulada
print(f"Cumulative explained variance ratio: {np.cumsum(pca.explained_variance_ratio_)}")
```

```python
# Ahora hacemos clustering con un Gaussian Mixture Model
from sklearn.mixture import GaussianMixture
# Creamos el modelo
gmm = GaussianMixture(n_components=3, random_state=42)
```

```python
# Calculamos el modelo
gmm.fit(X_pca)
# Predecimos los clusters
clusters = gmm.predict(X_pca)
```

```python
# Vemos los clusters
print(f"Clusters: {clusters}")
data['cluster'] = clusters
```

```python
# Ahora vemos con una grafica de pastel que tanto aporta cada cluster al USDAmount
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
clusters_name ={
    0: 'Personas Individuales',
    1: 'Parejas Sin Hijos',
    2: 'Familias Nucleares',
    3: 'Miscellaneous',
    4: 'Familias Grandes'
}
usdamount_by_cluster = data.groupby('cluster')['LocalCurrencyAmount'].sum()
labels = [clusters_name[c] for c in usdamount_by_cluster.index]
plt.pie(usdamount_by_cluster, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('DistribuciÃ³n de ingresos por Cluster')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(title='Clusters', loc='upper right')
plt.show()
```

```python
# Ahora veamos que tan facil es de predecir con una red neuronal de 2 capas para predecir el cluster
from sklearn.neural_network import MLPClassifier
# Creamos el modelo
nn_model = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=5, random_state=42)
# Separar los datos en entrenamiento y prueba
from sklearn.model_selection import train_test_split
# Hacerlo con los datos originales en vez de los PCA
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=['cluster', 'LocalCurrencyAmount']),
                                                    data['cluster'], test_size=0.2, random_state=42)
#X_train, X_test, y_train, y_test = train_test_split(X_pca, clusters, test_size=0.2, random_state=42)
# Entrenamos el modelo
nn_model.fit(X_train, y_train)
# Predecimos los clusters
y_pred = nn_model.predict(X_test)
# Vemos la precision del modelo
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Neural Network Model: {accuracy:.2f}")
```

```python
# Ahora el indice de silueta para ver que tan bien se separan los clusters
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(X_pca, clusters)
print(f"Average Silhouette Score: {silhouette_avg:.2f}")
```

```python
# Samplea los datosp para que sean el 10% de los datos originales para reducir el tiempo de ejecucion
sampled_data = data.sample(frac=0.2, random_state=42)
# Ahora sacamos el indice de silueta para los datos muestreados
silhouette_avg_sampled = silhouette_score(X_pca[sampled_data.index], sampled_data['cluster'])
print(f"Average Silhouette Score for Sampled Data: {silhouette_avg_sampled:.2f}")
```
